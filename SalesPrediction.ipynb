```python
import pandas as pd
```

```python
Sales = pd.read_csv("advertising.csv")
```

```python
Sales.head()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x = "TV", y = "Sales", data = Sales) 
```

```python
sns.scatterplot(x = "Radio", y = "Sales", data = Sales) 
```

```python
sns.scatterplot(x = "Newspaper", y = "Sales", data = Sales) 
```

```python
sns.distplot(Sales['Sales'])
```

```python
Sales['Sales'].max()
```

```python
sns.distplot(Sales['Sales'], hist = False)
```

```python
sns.distplot(Sales['Sales'], color = 'purple', kde = False, bins = 10)
```

As you can see in the above plot, the sales just got higher and higher and just before 10 and between 15 & 20 the sales got slighly constant and got higher just below 20 and again started falling off after 20 or just below 20.

```python
sns.jointplot(x = 'Sales', y = 'TV', data = Sales)
```

As per the above plot the sales is directly proportional to the TV i.e, the sales is increasing as the TV increases and vice-versa.

```python
sns.jointplot(x = 'Sales', y = 'Radio', data = Sales)
```

```python
sns.scatterplot(x = "Sales", y = "TV", hue = "TV", data = Sales) 
```

The Sales got high just below 15 and got higher and higher afterwards. 

```python
sns.scatterplot(x = "Radio", y = "Sales", hue = "Sales", data = Sales) 
```

The Sales is quite good in all ranges of Radio but is higher in case of Radio ranging from 30 to 50.

```python
sns.scatterplot(x = "TV", y = "Sales", hue = "Sales", data = Sales) 
```

As we analysed, the sales is directly proportional to TV, but highly intense for the TV ranging from 200 to 300.

```python
sns.scatterplot(x = "Newspaper", y = "Sales", hue = "Sales", data = Sales) 
```

The Sales is very much low between the newsapaper ranging from 60 to above 100. On the otherside the sales is pretty good for the newsapapers ranging below 60.
